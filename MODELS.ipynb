```python
from pythainlp import word_tokenize
from pythainlp.corpus import thai_stopwords
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer



#tokenization
lst_token = []

def get_word(text):
    tokens = word_tokenize(text,engine='newmm',keep_whitespace=False)
    return tokens
for t in range(len(df_test['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'])):
    word = get_word(df_test['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'][t])
    lst_token.append(word)

#add stopwords
stopwords = list(thai_stopwords())
stopwords.remove('ไม่')

remove_words = ['(',')','1.','2.','3.','4.','5','6','\u200b','“','”']
#                'แพ','แพง','โดยเฉพาะ','โยน','ใช้งาน','ไซน์','ไม่ต้อง','กลายเป็น',
#                'การเขียน','กี่','คล้ายๆ','ความพร้อม','คับ','คืด','คุ้มค่า','งานอดิเรก',
#                'จริง ๆ','จำนวนมาก','ซิ่ว','ชั้นนำ','ดัน','ดิ้ง','ดู','ตอนนี้',
#                'ตอนแรก','ตัดสินใจ','ตัวเอง','ติด','ถูกชะตา','ทดลอง','ทางใน','ตั้งใจ','เป็นลำดับ']
for i in range(len(remove_words)):
    stopwords.append(remove_words[i])

    
#remove stopwords
lst_not_in_stopwords = []
for i in range(len(lst_token)):
    list_word_not_stopwords = [j for j in lst_token[i] if j not in stopwords]
    lst_not_in_stopwords.append(list_word_not_stopwords)
lst_stw = []
for j in range(len(lst_not_in_stopwords)):
    list_no_stopwords = [k for k in lst_not_in_stopwords[j] if k != " "]
    lst_stw.append(list_no_stopwords)
    
    
#join list
reasons_lst = []
for i in range(len(lst_stw)):
    reasons_lst.append(''.join(lst_stw[i]))
    
    

#tfidf vectorizor
tvec = TfidfVectorizer(tokenizer=get_word, ngram_range=(1,3))
t_feat = tvec.fit_transform(reasons_lst)


#To dataframe
feature_names = tvec.get_feature_names()
corpus_index = [n for n in df_test['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง']]
df2 = pd.DataFrame(t_feat.todense(), index=corpus_index, columns=feature_names) ##result

```
