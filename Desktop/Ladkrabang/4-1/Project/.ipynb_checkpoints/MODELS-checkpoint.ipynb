```python
import pandas as pd


df_vector = pd.read_excel("vector1.xlsx", index_col=0)
#df_vector['Unnamed: 0'] = df_vector.index
df_vector.head(3)
```

```python
X = df_vector.drop(columns = ['target'])
y = df_vector['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
pd.DataFrame(y_test)
```

# MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
clf
```

```python
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
#https://coderzcolumn.com/tutorials/machine-learning/scikit-learn-sklearn-naive-bayes

from sklearn.model_selection import GridSearchCV
params = {'alpha': [0.01, 0.1, 0.5, 1.0, 10.0, ],}

multinomial_nb_grid = GridSearchCV(MultinomialNB(), param_grid=params, n_jobs=-1, cv=5, verbose=5)
multinomial_nb_grid.fit(X_train,y_train)

print('Train Accuracy : %.3f'%multinomial_nb_grid.best_estimator_.score(X_train, y_train))
print('Test Accuracy : %.3f'%multinomial_nb_grid.best_estimator_.score(X_test, y_test))
print('Best Accuracy Through Grid Search : %.3f'%multinomial_nb_grid.best_score_)
print('Best Parameters : ',multinomial_nb_grid.best_params_)
```

```python
multinomial_nb_grid.best_estimator_
```

```python
clf = MultinomialNB(alpha = 0.1).fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
#https://towardsdatascience.com/grid-search-for-parameter-tunning-3c6ff94e7a25

from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
scoring = {'accuracy': make_scorer(accuracy_score),
           'precision': make_scorer(precision_score, average = 'macro'),
           'recall': make_scorer(recall_score, average = 'macro'),
           'f1': make_scorer(f1_score, average = 'macro')}
#grid_search_rfc = GridSearchCV(rfc, param_grid = grid_values, scoring = scoring, refit='f1')
multinomial_nb_grid = GridSearchCV(MultinomialNB(), param_grid=params, scoring = scoring, refit='f1', n_jobs=-1, cv=5, verbose=5)
multinomial_nb_grid.fit(X_train, y_train)
```

```python
multinomial_nb_grid.best_estimator_
```

```python
clf = MultinomialNB(alpha=0.01).fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

# SVM

```python
from sklearn.svm import SVC

best_c = [10, 1.0, 0.1, 0.01]
gamma = [1, 0.1, 0.01, 0.001, 0.0001]
kernel = [ 'rbf', 'sigmoid','linear']

for c in best_c:
  for k in kernel:
    for g in gamma:
        clf = SVC(C=c, decision_function_shape = 'ovr', gamma =  g, random_state = 10, kernel = k)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = sum(y_test==y_pred)/len(y_test)
        print('C = ', c, '   kernel = ', k, '   gamma = ', g)
        print('   ', accuracy)

#radial basis function
```

```python
from sklearn.svm import SVC
param_grid = {'C': [0.01,0.1, 1, 10],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear','sigmoid']}  
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
print(grid.best_estimator_) 
```

```python
#C=1, gamma=0.0001, kernel=rbf, score=0.833
```

# RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
predicted= clf.predict(X_test)
print("RandomForest Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#https://towardsdatascience.com/grid-search-for-parameter-tunning-3c6ff94e7a25

rfc = RandomForestClassifier()
grid_values = {'n_estimators': [10, 30, 50, 100],
               'max_features': ['sqrt', 0.25, 0.5, 0.75, 1.0],
               'max_depth' : [4,5,6,7,8]
              }
grid_search_rfc = GridSearchCV(rfc, param_grid = grid_values, scoring = 'accuracy',verbose=5)
grid_search_rfc.fit(X_train, y_train)
```

```python
grid_search_rfc.best_params_
```

```python
n_estimators = [10, 30, 50, 100,1000]
max_features = ['sqrt', 0.25, 0.5, 0.75, 1.0]
max_depth = [4,5,6,7,8]

for n in n_estimators:
  for f in max_features:
    for d in max_depth:
        clf = RandomForestClassifier(max_depth = d, max_features = f, n_estimators=n, random_state=10)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = sum(y_test==y_pred)/len(y_test)
        print('max_depth = ', d, '   max_features = ', f, '   n_estimators = ', n)
        print('   ', accuracy)
```

```python
classifier = RandomForestClassifier(max_depth = 6, max_features = 0.75 ,n_estimators=10, random_state=10)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
predicted= clf.predict(X_test)
print("RandomForest Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 8)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print("KNN:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

error = []
range_k = range(1,30)

for k in range_k:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))

plt.figure(figsize=(12, 6))
plt.plot(range(1, 30), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
best_k = range_k[error.index(min(error))]
print("The best k is " ,best_k)

#https://stackabuse.com/k-nearest-neighbors-algorithm-in-python-and-scikit-learn/
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("KNN:",metrics.accuracy_score(y_test, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```
