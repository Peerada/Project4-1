```python
from pythainlp import word_tokenize  

lst_token = []

def get_word(text):
    tokens = word_tokenize(text,engine='newmm',keep_whitespace=False)
    return tokens

for t in range(len(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'])):
    word = get_word(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'][t])
    lst_token.append(word)
    #for i in range(len(X)):
        #word = get_word(X[X.columns[t]][i])
    #test = [i for i in word if i not in x]
    #test
```

```python
lst_token
```

```python
#https://python3.wannaphong.com/2020/02/ai.html

from pythainlp.corpus import thai_stopwords
stopwords = list(thai_stopwords())
stopwords.remove('ไม่')

remove_words = ['(',')','1.','2.','3.','4.','5','6','\u200b','“','”']
#                'แพ','แพง','โดยเฉพาะ','โยน','ใช้งาน','ไซน์','ไม่ต้อง','กลายเป็น',
#                'การเขียน','กี่','คล้ายๆ','ความพร้อม','คับ','คืด','คุ้มค่า','งานอดิเรก',
#                'จริง ๆ','จำนวนมาก','ซิ่ว','ชั้นนำ','ดัน','ดิ้ง','ดู','ตอนนี้',
#                'ตอนแรก','ตัดสินใจ','ตัวเอง','ติด','ถูกชะตา','ทดลอง','ทางใน','ตั้งใจ','เป็นลำดับ']
for i in range(len(remove_words)):
    stopwords.append(remove_words[i])
stopwords
```

```python
lst_not_in_stopwords = []
for i in range(len(lst_token)):
    list_word_not_stopwords = [j for j in lst_token[i] if j not in stopwords]
    lst_not_in_stopwords.append(list_word_not_stopwords)
lst_stw = []
for j in range(len(lst_not_in_stopwords)):
    list_no_stopwords = [k for k in lst_not_in_stopwords[j] if k != " "]
    lst_stw.append(list_no_stopwords)
lst_stw
```

```python
from collections import OrderedDict

lst_exp_1 = ['C','C++',"C#","Java"]
lst_exp_2 = ["Python"]
lst_exp_3 = ["Web (CSS/HTML)", "JavaScript", "PHP"]
lst_exp_4 = ["dart"]
lst_exp_5 = ["adobe dreamweaver", "กราฟฟิค", "การตัดต่อภาพ", "การตัดต่อวีดีโอ"]
lst_exp_6 = ["ไม่มีพื้นฐานอะไรเลย"]

lst_all = []
for i in range(len(df_exp)):
    value = df_exp['new'].iloc[i]
    lst_exp = []
    for j in range(len(value)):
        words = value[j]
        if words in lst_exp_1 :
            lst_exp.append('Programming')
        if words in lst_exp_2 :
            lst_exp.append('Python')
        if words in lst_exp_3 :
            lst_exp.append('Web Programming')
        if words in lst_exp_4 :
            lst_exp.append('App Programing')
        if words in lst_exp_5 :
            lst_exp.append('Graphic')
        if words in lst_exp_6 :
            lst_exp.append('No exp')
    
    lst_exp = list(OrderedDict.fromkeys(lst_exp))
    print(lst_exp)
    lst_all.append(lst_exp)
        #column_values = pd.Series(lst_exp)
        #df_exp.insert(loc=i, column='group', value=column_values)
```

```python
import pandas as pd


df_vector = pd.read_excel("vector.xlsx", index_col=0)
#df_vector['Unnamed: 0'] = df_vector.index
df_vector
```

```python
X = df_vector
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
clf
```

```python
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'alpha': [0.01, 0.1, 0.5, 1.0, 10.0, ],}

multinomial_nb_grid = GridSearchCV(MultinomialNB(), param_grid=params, n_jobs=-1, cv=5, verbose=5)
multinomial_nb_grid.fit(X_train,y_train)

print('Train Accuracy : %.3f'%multinomial_nb_grid.best_estimator_.score(X_train, y_train))
print('Test Accuracy : %.3f'%multinomial_nb_grid.best_estimator_.score(X_test, y_test))
print('Best Accuracy Through Grid Search : %.3f'%multinomial_nb_grid.best_score_)
print('Best Parameters : ',multinomial_nb_grid.best_params_)
```

```python
multinomial_nb_grid.best_estimator_
```

```python

clf = MultinomialNB(alpha=1.0).fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
#https://towardsdatascience.com/grid-search-for-parameter-tunning-3c6ff94e7a25

from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
scoring = {'accuracy': make_scorer(accuracy_score),
           'precision': make_scorer(precision_score, average = 'macro'),
           'recall': make_scorer(recall_score, average = 'macro'),
           'f1': make_scorer(f1_score, average = 'macro')}
#grid_search_rfc = GridSearchCV(rfc, param_grid = grid_values, scoring = scoring, refit='f1')
multinomial_nb_grid = GridSearchCV(MultinomialNB(), param_grid=params, scoring = scoring, refit='f1', n_jobs=-1, cv=5, verbose=5)
multinomial_nb_grid.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC

best_c = [50, 10, 1.0, 0.1, 0.01]
gamma: [1, 0.1, 0.01, 0.001, 0.0001]
kernel = ['poly', 'rbf', 'sigmoid','linear']

for c in best_c:
  for k in kernel:
    clf = SVC(C=c, decision_function_shape = 'ovr', gamma =  gamma, random_state = 10, kernel = k)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = sum(y_test==y_pred)/len(y_test)
    print('C = ', c, '   kernel = ', k)
    print('   ', accuracy)
```
