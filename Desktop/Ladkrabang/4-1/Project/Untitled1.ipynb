```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx")
```

```python
df = df.drop(columns=['รุ่น','ระดับความชอบในการเรียนคณะไอทีลาดกระบัง','ชั้นปี','สาขา',
                        'แขนง','รอบการสมัคร','ถ้าเคยคิดจะซิ่ว จะซิ่วเพราะอะไร',
                        'หลังจากเข้ามาเรียนแล้วความรู้สึกเปลี่ยนไปจากตอนแรกไหม เพราะอะไร',
                       'เคยคิดจะซิ่วไหม'])
df = df.head(5)
```

```python
df
```

```python
from pythainlp import word_tokenize  

lst_token = []

def get_word(text):
    tokens = word_tokenize(text,engine='newmm',keep_whitespace=False)
    return tokens

for t in range(len(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'])):
    word = get_word(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'][t])
    lst_token.append(word)
    #for i in range(len(X)):
        #word = get_word(X[X.columns[t]][i])
    #test = [i for i in word if i not in x]
    #test
```

```python
from pythainlp.corpus import thai_stopwords
stopwords = list(thai_stopwords())
list_word_not_stopwords = [i for i in lst_token if i not in stopwords]
print(list_word_not_stopwords)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tokens_list_j = [','.join(tkn) for tkn in lst_token]
tvec = TfidfVectorizer(analyzer=lambda x:x.split(','), stop_words = thai_stopwords())
t_feat = tvec.fit_transform(tokens_list_j)
```

```python
cvec = CountVectorizer(analyzer=lambda x:x.split(','))
c_feat = cvec.fit_transform(tokens_list_j)
```

```python
vec = pd.DataFrame(c_feat.todense(), columns=cvec.vocabulary_)
vec
```
