```python
import pandas as pd
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV


df_vector = pd.read_excel("vector1.xlsx", index_col=0)
#df_vector['Unnamed: 0'] = df_vector.index
df_vector.head(3)
```

```python
X = df_vector.drop(columns = ['target',
                             'แพ','แพง','โดยเฉพาะ','โยน','ใช้งาน','ไซน์','ไม่ต้อง','กลายเป็น',
                             'การเขียน','กี่','คล้ายๆ','ความพร้อม','คับ','คืด','คุ้มค่า','งานอดิเรก',
                             'จริง ๆ','จำนวนมาก','ซิ่ว','ชั้นนำ','ดัน','ดิ้ง','ดู','ตอนนี้',
                             'ตอนแรก','ตัดสินใจ','ตัวเอง','ติด','ถูกชะตา','ทดลอง','ทางใน','ตั้งใจ'])
#X = df_vector.drop(columns = ['target'])
y = df_vector['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC
clf = SVC(C=0.01, gamma=1, kernel='linear',random_state = None)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = sum(y_test==y_pred)/len(y_test)
#print('C = ', c, '   kernel = ', k)
print('accuracy = ', accuracy)
```

```python
from sklearn.svm import SVC

best_c = [10, 1.0, 0.1, 0.01]
gamma = [1, 0.1, 0.01, 0.001, 0.0001]
kernel = [ 'rbf', 'sigmoid','linear']

for c in best_c:
  for k in kernel:
    for g in gamma:
        clf = SVC(C=c, decision_function_shape = 'ovr', gamma =  g, random_state = 10, kernel = k)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = sum(y_test==y_pred)/len(y_test)
        print('C = ', c, '   kernel = ', k, '   gamma = ', g)
        print('   ', accuracy)
```

```python
n_estimators = [10, 30, 50, 100,1000]
max_features = ['sqrt', 0.25, 0.5, 0.75, 1.0]
max_depth = [4,5,6,7,8]

for n in n_estimators:
  for f in max_features:
    for d in max_depth:
        clf = RandomForestClassifier(max_depth = d, max_features = f, n_estimators=n, random_state=10)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = sum(y_test==y_pred)/len(y_test)
        print('max_depth = ', d, '   max_features = ', f, '   n_estimators = ', n)
        print('   ', accuracy)
```
