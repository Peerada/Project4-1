import re
import string

def clean_msg(msg):
    # ลบ เครื่องหมายคำพูด (punctuation)
    for c in string.punctuation:
        msg = re.sub(r'\{}'.format(c),'',msg)
    return msg

print('original text:\n',df)
print('clean text:\n',clean_msg(str(df)))

```python
lst = []
for i in df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง']:
    lst.append(i)
lst_text = []
for i in range(len(lst)):
    list_word = word_tokenize(lst[i])
    lst_text.append(list_word)
```

```python
lst_text
```

```python
#https://python3.wannaphong.com/2020/02/ai.html
stopwords = list(thai_stopwords())
stopwords.remove('ไม่')
stopwords
```

```python
lst_not_in_stopwords = []
for i in range(len(lst_text)):
    list_word_not_stopwords = [j for j in lst_text[i] if j not in stopwords]
    lst_not_in_stopwords.append(list_word_not_stopwords)
lst_stw = []
for j in range(len(lst_not_in_stopwords)):
    list_no_stopwords = [k for k in lst_not_in_stopwords[j] if k != " "]
    lst_stw.append(list_no_stopwords)
```

```python
y = df['ระดับความชอบในการเรียนคณะไอทีลาดกระบัง']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
