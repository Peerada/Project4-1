```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("data.xlsx")
df = df.drop([38])
df = df.drop(columns=['รุ่น','ระดับความชอบในการเรียนคณะไอทีลาดกระบัง','ชั้นปี','สาขา',
                        'แขนง','รอบการสมัคร','ถ้าเคยคิดจะซิ่ว จะซิ่วเพราะอะไร',
                        'หลังจากเข้ามาเรียนแล้วความรู้สึกเปลี่ยนไปจากตอนแรกไหม เพราะอะไร',
                       'เคยคิดจะซิ่วไหม'])
```

```python
from pythainlp import word_tokenize  

lst_token = []

def get_word(text):
    tokens = word_tokenize(text,engine='newmm',keep_whitespace=False)
    return tokens

for t in range(len(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'])):
    word = get_word(df['เหตุผลที่เลือกเรียนคณะไอทีลาดกระบัง'][t])
    lst_token.append(word)
    #for i in range(len(X)):
        #word = get_word(X[X.columns[t]][i])
    #test = [i for i in word if i not in x]
    #test
```

```python
#https://python3.wannaphong.com/2020/02/ai.html

from pythainlp.corpus import thai_stopwords
stopwords = list(thai_stopwords())
stopwords.remove('ไม่')

remove_words = ['(',')','1.','2.','3.','4.','5','6','\u200b','“','”',',']
#                'แพ','แพง','โดยเฉพาะ','โยน','ใช้งาน','ไซน์','ไม่ต้อง','กลายเป็น',
#                'การเขียน','กี่','คล้ายๆ','ความพร้อม','คับ','คืด','คุ้มค่า','งานอดิเรก',
#                'จริง ๆ','จำนวนมาก','ซิ่ว','ชั้นนำ','ดัน','ดิ้ง','ดู','ตอนนี้',
#                'ตอนแรก','ตัดสินใจ','ตัวเอง','ติด','ถูกชะตา','ทดลอง','ทางใน','ตั้งใจ','เป็นลำดับ']
for i in range(len(remove_words)):
    stopwords.append(remove_words[i])
stopwords
```

```python
lst_not_in_stopwords = []
for i in range(len(lst_token)):
    list_word_not_stopwords = [j for j in lst_token[i] if j not in stopwords]
    lst_not_in_stopwords.append(list_word_not_stopwords)
lst_stw = []
for j in range(len(lst_not_in_stopwords)):
    list_no_stopwords = [k for k in lst_not_in_stopwords[j] if k != " "]
    lst_stw.append(list_no_stopwords)
lst_stw
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tokens_list_j = [','.join(tkn) for tkn in lst_stw]

#COUNTVEC
cvec = CountVectorizer(analyzer=lambda x:x.split(','))
c_feat = cvec.fit_transform(tokens_list_j)



#TFIDF
tvec = TfidfVectorizer(tokenizer=get_word, ngram_range=(1,3))
t_feat = tvec.fit_transform(tokens_list_j)
```

```python
#tvec.vocabulary_
full_vocab = tvec.get_feature_names()

#สองตัวข้างบนได้ผลลัพธ์ออกมาเหมือนกันและจำนวนเท่ากัน แต่เรียงลำดับไม่เหมือนกัน และvocabularyจะแสดงidของคำนั้นๆด้วย
```
